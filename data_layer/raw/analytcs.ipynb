## Fase 2: Visão de Qualidade

### 2.1. Detecção de Valores Ausentes
Quantificamos o número e o percentual de valores nulos (`NaN`) por coluna.

```python
nulos = df.isnull().sum()
percentual_nulos = (nulos / len(df)) * 100

df_nulos = pd.DataFrame({
    'Total Nulos': nulos,
    'Percentual (%)': percentual_nulos
})

# Exibir apenas colunas que possuem valores nulos, ordenadas pela mais crítica
df_nulos = df_nulos[df_nulos['Total Nulos'] > 0].sort_values(by='Percentual (%)', ascending=False)

if df_nulos.empty:
    print(" Não há valores ausentes (nulos) no dataset.")
else:
    print("--- Detecção de Valores Ausentes ---")
    display(df_nulos)
```

### 2.2. Verificação de Dados Duplicados
Identificamos e contamos a quantidade de linhas **completamente** duplicadas. 

```python
duplicadas = df.duplicated().sum()
percentual_duplicadas = (duplicadas / len(df)) * 100

print(f"Total de linhas completamente duplicadas: {duplicadas}")
print(f"Percentual de linhas duplicadas: {percentual_duplicadas:.2f}%")
```

### 2.3. Análise de Consistência e Intervalo
Para colunas categóricas, verificamos os valores únicos para identificar inconsistências (ex: "Brasil" vs "brazil"). Para numéricas, o `.describe()` ajuda a ver intervalos (ex: valores negativos em idade).

```python
print("--- Verificação de Consistência (Categóricas) ---")
# Selecionar colunas categóricas (object ou category)
colunas_categoricas = df.select_dtypes(include=['object', 'category']).columns

if colunas_categoricas.empty:
    print("Nenhuma coluna categórica (object/category) encontrada.")
else:
    for col in colunas_categoricas:
        print(f"\nValores únicos na coluna '{col}' (Top 50 mais frequentes):")
        print(df[col].value_counts().head(50).index.to_list())
        
        num_unicos = df[col].nunique()
        if num_unicos > 50:
            print(f"Atenção: Coluna '{col}' possui {num_unicos} valores únicos (alta cardinalidade).")
        else:
            print(f"Total de valores únicos: {num_unicos}")
```

```python
print("--- Verificação de Intervalo (Numéricas) ---")
# .describe() para ver min/max e identificar valores inesperados
colunas_numericas = df.select_dtypes(include=np.number).columns

if colunas_numericas.empty:
    print("Nenhuma coluna numérica encontrada.")
else:
    print("Analisar 'min' e 'max' para valores fora do intervalo esperado (ex: idade negativa).")
    display(df[colunas_numericas].describe())
```

### 2.5. Identificação de Outliers
Utilizamos Boxplots (diagramas de caixa) para identificar visualmente outliers nas variáveis numéricas.

```python
print("--- Gerando Boxplots para Detecção de Outliers ---")
colunas_numericas = df.select_dtypes(include=np.number).columns

if len(colunas_numericas) > 0:
    print(f"Gerando Boxplots para: {colunas_numericas.to_list()}")
    n_cols_grid = 3
    n_rows_grid = (len(colunas_numericas) - 1) // n_cols_grid + 1
    
    plt.figure(figsize=(n_cols_grid * 5, n_rows_grid * 4))
    
    for i, col in enumerate(colunas_numericas):
        plt.subplot(n_rows_grid, n_cols_grid, i + 1)
        sns.boxplot(y=df[col])
        plt.title(f'Boxplot de {col}')
        plt.ylabel(col)
    
    plt.tight_layout()
    plt.show()
else:
    print("Nenhuma coluna numérica encontrada para gerar boxplots.")
```

## Fase 3: Visão Estatística (Análise Descritiva e Padrões)

### 3.1 Estatísticas Descritivas

```python
print("--- Frequência (Categóricas) - Top 10 por coluna ---")
colunas_categoricas = df.select_dtypes(include=['object', 'category']).columns

if colunas_categoricas.empty:
    print("Nenhuma coluna categórica encontrada.")
else:
    for col in colunas_categoricas:
        print(f"\n--- Frequência da Coluna: {col} ---")
        # Exibe o Top 10 em percentual
        display(df[col].value_counts(normalize=True).head(10) * 100)
```

### 3.2. Análise de Distribuição
Criamos histogramas (com KDE - linha de densidade) para visualizar a distribuição das variáveis numéricas e gráficos de barras para as categóricas.

```python
print("--- Gerando Histogramas (Distribuição Numéricas) ---")
colunas_numericas = df.select_dtypes(include=np.number).columns

if len(colunas_numericas) > 0:
    n_cols_grid = 3
    n_rows_grid = (len(colunas_numericas) - 1) // n_cols_grid + 1
    
    plt.figure(figsize=(n_cols_grid * 5, n_rows_grid * 4))
    
    for i, col in enumerate(colunas_numericas):
        plt.subplot(n_rows_grid, n_cols_grid, i + 1)
        sns.histplot(df[col], kde=True, bins=30)
        plt.title(f'Distribuição de {col}')
    
    plt.tight_layout()
    plt.show()
else:
    print("Nenhuma coluna numérica para histogramas.")
```

```python
print("--- Gerando Gráficos de Barras (Frequência Categóricas) ---")
colunas_categoricas = df.select_dtypes(include=['object', 'category']).columns

# Filtrar colunas com muitas categorias únicas para não poluir os gráficos
col_cat_plot = [col for col in colunas_categoricas if df[col].nunique() < 50]

if len(col_cat_plot) > 0:
    n_cols_grid = 2
    n_rows_grid = (len(col_cat_plot) - 1) // n_cols_grid + 1
    
    plt.figure(figsize=(n_cols_grid * 7, n_rows_grid * 5))
    
    for i, col in enumerate(col_cat_plot):
        plt.subplot(n_rows_grid, n_cols_grid, i + 1)
        # Ordenar barras pela frequência 
        order = df[col].value_counts().index
        sns.countplot(data=df, y=col, order=order)
        plt.title(f'Frequência de {col}')
    
    plt.tight_layout()
    plt.show()
else:
    print("Nenhuma coluna categórica (com < 50 valores únicos) encontrada para gerar gráficos de barras.")
```

### 3.3. Análise de Relações e Correlações
Calculamos a matriz de correlação (usando o método de Pearson) e a visualizamos com um Mapa de Calor (Heatmap) para identificar relações lineares entre variáveis numéricas.

Em seguida, usamos um `pairplot` para visualizar gráficos de dispersão (scatter plots) entre pares de variáveis, ajudando a identificar relações não lineares.

```python
print("--- Gerando Mapa de Calor (Correlação de Pearson) ---")
df_numeric = df.select_dtypes(include=np.number)

if df_numeric.shape[1] > 1:
    corr_matrix = df_numeric.corr(method='pearson')
    
    plt.figure(figsize=(12, 8))
    sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
    plt.title('Matriz de Correlação (Pearson)')
    plt.show()  
else:
    print("Não há colunas numéricas suficientes (>1) para gerar uma matriz de correlação.")
```

```python
print("--- Gerando Pairplot (Relação entre Pares Numéricos) ---")

if 'df_numeric' in locals() and df_numeric.shape[1] > 1:
    df_sample = df
    
    if len(df) > 5000:
        print("Dataset grande detectado. Usando amostra aleatória de 5000 linhas para o pairplot.")
        df_sample = df.sample(5000, random_state=42)

    cols_to_plot = df_numeric.columns
    sns.pairplot(df_sample[cols_to_plot], diag_kind='kde')
    plt.show()

else:
    print("Não há colunas numéricas suficientes (>1) para gerar um pairplot.")
```

### 3.4. Análise Temporal
Buscamos colunas de data/hora (`datetime`) para agregar os dados por períodos (ex: Mês) e identificar tendências ou sazonalidade na contagem de registros.

```python
print("--- Análise Temporal (Tendência de Registros) ---")

colunas_data = df.select_dtypes(include=['datetime', 'datetime64[ns]']).columns
df_temporal = df.copy() 

# Realiza a análise se uma coluna de data foi encontrada
if len(colunas_data) > 0:
    col_data_principal = colunas_data[0]
    print(f"Usando coluna '{col_data_principal}' como eixo de tempo principal.")
    try:
        df_temporal_idx = df_temporal.set_index(col_data_principal)
        eventos_por_mes = df_temporal_idx.resample('M').size()
        eventos_por_mes.name = 'Contagem de Registros'
        
        if eventos_por_mes.empty or eventos_por_mes.nunique() == 1:
                print(f"Os dados da coluna '{col_data_principal}' parecem estar em um período muito curto para análise mensal.")
        else:
            plt.figure(figsize=(12, 6))
            eventos_por_mes.plot(title='Tendência Mensal (Contagem de Registros)', marker='o')
            plt.ylabel('Contagem de Registros')
            plt.xlabel('Mês')
            plt.show()
            
    except Exception as e:
        print(f"Não foi possível agregar os dados temporalmente. Erro: {e}")
else:
    print("Nenhuma coluna temporal (datetime ou convertível) foi encontrada para análise.")
```

## Fase 4: Visão de Negócio

### 4.1. Cálculo de Métricas de Negócio
Cálculo de métricas básicas como:
- Métricas de Faturamento e Distância
- Métricas de Sucesso vs. Falha (Cancelamento)
- Métricas de Avaliação (Ratings)

```python
print("--- 4.1. Cálculo de Métricas de Negócio ---")

# --- Métricas de Faturamento e Distância ---
try:
    total_bookings = len(df)
    print(f"Total de Registros (Bookings) no dataset: {total_bookings}")

    # Faturamento só calculado sobre corridas completas
    df_completed = df[df['Booking Status'] == 'Completed']
    
    total_faturamento = df_completed['Booking Value'].sum()
    media_faturamento = df_completed['Booking Value'].mean()
    media_distancia = df_completed['Ride Distance'].mean()

    print(f"\nFaturamento Total (de corridas 'Completed'): {total_faturamento:,.2f}")
    print(f"Valor Médio (de corridas 'Completed'): {media_faturamento:,.2f}")
    print(f"Distância Média (de corridas 'Completed'): {media_distancia:,.2f} km") 

except KeyError as e:
    print(f"Erro: Coluna não encontrada para métricas de faturamento: {e}")
except Exception as e:
    print(f"Erro ao calcular métricas: {e}")


# --- Métricas de Sucesso vs. Falha (Cancelamento) ---
try:
    if total_bookings > 0:
        count_completed = df[df['Booking Status'] == 'Completed'].shape[0]
        count_cancel_driver = df[df['Booking Status'] == 'Cancelled by Driver'].shape[0]
        count_cancel_customer = df[df['Booking Status'] == 'Cancelled by Customer'].shape[0]
        count_no_driver = df[df['Booking Status'] == 'No Driver Found'].shape[0]
        count_incomplete = df[df['Booking Status'] == 'Incomplete'].shape[0]
        
        total_failures = count_cancel_driver + count_cancel_customer + count_no_driver + count_incomplete
        
        taxa_sucesso = (count_completed / total_bookings) * 100
        taxa_falha_geral = (total_failures / total_bookings) * 100
        taxa_cancel_driver = (count_cancel_driver / total_bookings) * 100
        taxa_cancel_customer = (count_cancel_customer / total_bookings) * 100
        taxa_no_driver = (count_no_driver / total_bookings) * 100
        taxa_incomplete = (count_incomplete / total_bookings) * 100

        print(f"\n--- Métricas de Status da Corrida ---")
        print(f"Taxa de Sucesso ('Completed'): {taxa_sucesso:.2f}%  (Total: {count_completed})")
        print(f"Taxa de Falha (Geral): {taxa_falha_geral:.2f}%  (Total: {total_failures})")
        
        print(f"\n--- Detalhamento das Falhas ---")
        print(f"  - Cancelled by Driver: {taxa_cancel_driver:.2f}%  (Total: {count_cancel_driver})")
        print(f"  - Cancelled by Customer: {taxa_cancel_customer:.2f}%  (Total: {count_cancel_customer})")
        print(f"  - No Driver Found: {taxa_no_driver:.2f}%  (Total: {count_no_driver})")
        print(f"  - Incomplete: {taxa_incomplete:.2f}%  (Total: {count_incomplete})")

        soma_verificacao = count_completed + total_failures
        print(f"\nVerificação (Total Contado): {soma_verificacao} (deve ser igual a {total_bookings})")

except KeyError as e:
    print(f"Erro: Coluna 'Booking Status' não encontrada: {e}")
except Exception as e:
    print(f"Erro ao calcular status: {e}")


# --- Métricas de Avaliação (Ratings) ---
try:
    media_rating_motorista = df_completed['Driver Ratings'].mean()
    media_rating_cliente = df_completed['Customer Rating'].mean()
    print(f"\n--- Métricas de Avaliação (de corridas 'Completed') ---")
    print(f"Nota Média dos Motoristas: {media_rating_motorista:.2f}")
    print(f"Nota Média dos Clientes: {media_rating_cliente:.2f}")
except KeyError as e:
    print(f"Erro: Coluna não encontrada para métricas de rating: {e}")
```

### 4.2. Análise de Segmentação
Agrupamos os dados por segmentos de métodos de pagamento.

```python
print("--- 4.2.2. Análise de Segmentação por 'Payment Method' ---")

COLUNA_SEGMENTO = 'Payment Method'
COLUNA_METRICA_VALOR = 'Booking Value'

try:
    # Agrupar por método de pagamento e calcular a SOMA (Total Faturado)
    valor_total_por_pagamento = df.groupby(COLUNA_SEGMENTO)[COLUNA_METRICA_VALOR].sum().sort_values(ascending=False)
    
    print(f"\n--- Faturamento Total (Soma) por {COLUNA_SEGMENTO} ---")
    print(valor_total_por_pagamento)
    
    plt.figure(figsize=(10, 5))
    valor_total_por_pagamento.plot(kind='bar', color='lightgreen')
    plt.title(f'Faturamento Total ({COLUNA_METRICA_VALOR}) por {COLUNA_SEGMENTO}')
    plt.ylabel('Valor Total Faturado')
    plt.xlabel('Método de Pagamento')
    plt.xticks(rotation=45, ha='right')
    plt.show()

except KeyError:
    print(f"Erro: Colunas '{COLUNA_SEGMENTO}' ou '{COLUNA_METRICA_VALOR}' não encontradas.")
```

### 4.3. Identificação de Padrões Relevantes
Identificamos os itens mais frequentes (Top 10) em colunas de alta relevância para o negócio.

```python
print("--- 4.3. Identificação de Padrões Relevantes (Top Frequentes) ---")

# --- Top 10 Locais de Partida (Pickup) ---
try:
    COLUNA_PICKUP = 'Pickup Location'
    print(f"\n--- Top 10 Locais de Partida ({COLUNA_PICKUP}) ---")
    top_pickups = df[COLUNA_PICKUP].value_counts().head(10)
    display(top_pickups)
except KeyError:
    print(f"Erro: Coluna '{COLUNA_PICKUP}' não encontrada.")


# --- Top 10 Locais de Chegada (Drop) ---
try:
    COLUNA_DROP = 'Drop Location'
    print(f"\n--- Top 10 Locais de Chegada ({COLUNA_DROP}) ---")
    top_drops = df[COLUNA_DROP].value_counts().head(10)
    display(top_drops)
except KeyError:
    print(f"Erro: Coluna '{COLUNA_DROP}' não encontrada.")


# --- Top 5 Motivos de Cancelamento (Cliente) ---
try:
    COLUNA_MOTIVO_CLIENTE = 'Reason for cancelling by Customer'
    print(f"\n--- Top 5 Motivos de Cancelamento (Cliente) ---")
    # .dropna() é importante aqui pois a coluna tem muitos nulos
    top_motivos_cliente = df[COLUNA_MOTIVO_CLIENTE].value_counts().head(5)
    display(top_motivos_cliente)
except KeyError:
    print(f"Erro: Coluna '{COLUNA_MOTIVO_CLIENTE}' não encontrada.")


# --- Top 5 Motivos de Cancelamento (Motorista) ---
try:
    COLUNA_MOTIVO_MOTORISTA = 'Driver Cancellation Reason'
    print(f"\n--- Top 5 Motivos de Cancelamento (Motorista) ---")
    top_motivos_motorista = df[COLUNA_MOTIVO_MOTORISTA].value_counts().head(5)
    display(top_motivos_motorista)
except KeyError:
    print(f"Erro: Coluna '{COLUNA_MOTIVO_MOTORISTA}' não encontrada.")
```

## Fase 5: Visão de Governança 

### 5.1. Criação do Dicionário de Dados
Geramos um resumo (Data Dictionary) do dataset final, incluindo tipo de dado, contagem de nulos e um espaço para descrição manual. Este resumo também é salvo em `dicionario_de_dados.csv`.

```python
print("--- 5.1. Dicionário de Dados ---")

# Coletar informações
col_names = df.columns
col_types = [str(dtype) for dtype in df.dtypes]
col_non_nulls = df.count()
col_nulls = df.isnull().sum()
col_null_percent = (col_nulls / len(df)) * 100
col_uniques = df.nunique()

# Criar o Dicionário
dicionario_dados = pd.DataFrame({
    'Coluna': col_names,
    'Tipo de Dado': col_types,
    'Valores Únicos': col_uniques,
    'Valores Não-Nulos': col_non_nulls,
    'Valores Nulos': col_nulls,
    'Percentual Nulos (%)': col_null_percent.round(2),
    'Descrição (Preencher Manualmente)': ['' for _ in col_names] 
})

# Resetar o índice para que 'Coluna' seja uma coluna normal
dicionario_dados = dicionario_dados.reset_index(drop=True)

# Exibir o dicionário completo na tela
print("Dicionário de Dados Gerado:")
display(dicionario_dados)
```
